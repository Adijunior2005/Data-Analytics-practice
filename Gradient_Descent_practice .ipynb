```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ma=[]
ba=[]
x=np.array([1,2,3,5,6,4,8])
y=np.array([4,5,3,6,7,8,5])
def gradient_descent(x,y):
    m=b=0
    n=len(x)
    iterations=1000
    learning_rate=0.0001
    for i in range(iterations):
        yp=m*x+b
        cost=(1/n)*sum((val**2 for val in (yp)))
        md=-(2/n)*sum(x*(y-yp))
        bd=-(2/n)*sum(y-yp)
        m=m-learning_rate*md
        b=b-learning_rate*bd
        print("m {},b {},cost {},iteration {}".format(m,b,cost,i))
        ma.append(m)
        ba.append(b)
gradient_descent(x,y)
for i in range(0, len(ma), 2):
    plt.plot(ma[i:i+2], ba[i:i+2], '-o')
    
plt.show() 
    
```
